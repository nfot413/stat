```python
import numpy as np
import matplotlib.pyplot as plt

k_B = 1.38e-23   # 玻尔兹曼常数
Delta_E = 1e-21  # 激发态能量差

# 温度范围 (K)
T = np.linspace(1, 1000, 500)

P0 = 1 / (1 + np.exp(-Delta_E / (k_B * T)))  # 基态概率
P1 = 1 - P0                                  # 激发态概率

# 绘图
plt.plot(T, P0, label="Ground state (E0)")
plt.plot(T, P1, label="Excited state (E1)")
plt.xlabel("Temperature T (K)")
plt.ylabel("Probability P")
plt.title("Boltzmann Distribution (Two-level system)")
plt.legend()
plt.show()
```

## 玻尔兹曼分布

上图简单展示了一个双能级系统随温度变化的概率分布

其数学表达式为
$$
P(E_0) = \frac{1}{1 + e^{-\Delta E / (k_B T)}}, 
\quad
P(E_1) = \frac{e^{-\Delta E / (k_B T)}}{1 + e^{-\Delta E / (k_B T)}}
$$
